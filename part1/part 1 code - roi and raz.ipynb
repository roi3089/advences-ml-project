```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

def parse_car_page(url):
    #      DOM  BeautifulSoup
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #    
    data = {
        'manufactor': 'Renault',  #     
        'Year': None,
        'model': None,
        'Hand': None,
        'Gear': None,
        'Engine_capacity': None,
        'Engine_type': None,
        'Prev_ownership': None,
        'Curr_ownership': None,
        'Area': None,
        'City': None,
        'Price': None,
        'Pic_num': None,
        'Cre_date': None,
        'Repub_date': None,
        'Description': None,
        'Color': None,
        'Km': None,
        'Test': None,
        'Supply_score': None  #     ,    
    }
    
    #   
    price_and_model_div = soup.find('div', class_='d-flex justify-content-between')
    if price_and_model_div:
        h2_tags = price_and_model_div.find_all('h2', class_='card-title')
        if len(h2_tags) == 2:
            data['model'] = h2_tags[0].text.strip()  #  
            price_text = h2_tags[1].text.strip().replace('', '').replace(',', '').strip()
            data['Price'] = float(price_text) if price_text.isdigit() else None  # 
    
    #   
    pic_div = soup.find('div', itemtype='http://schema.org/ImageGallery')
    if pic_div:
        data['Pic_num'] = len(pic_div.find_all('figure'))  #  
    
    #  
    description_p = soup.find('p', class_='text-word-break')
    if description_p:
        data['Description'] = description_p.text.strip()  # 
    
    #    
    dates_div = soup.find('div', class_='d-flex flex-row align-items-center justify-content-center flex-wrap')
    if dates_div:
        date_texts = dates_div.find_all('div', class_='px-3')
        for date_text in date_texts:
            text = date_text.text.strip()
            if ' ' in text:
                data['Cre_date'] = datetime.strptime(text.replace(' : ', '').strip(), '%d/%m/%Y')  #  
            elif '  ' in text:
                data['Repub_date'] = datetime.strptime(text.replace('  : ', '').strip(), '%d/%m/%Y')  #  
    
    #   
    rows = soup.find('table', class_='table').find_all('tr')
    for row in rows:
        cells = row.find_all('td')
        if len(cells) == 2:
            label = cells[0].text.strip()
            value = cells[1].text.strip()
            
            #    
            if label == '':
                data['Year'] = int(value)  # 
            elif label == '':
                data['Hand'] = int(value)  # 
            elif label == '. ':
                data['Gear'] = value  #  
            elif label == '':
                data['Engine_capacity'] = int(value.replace(',', ''))  #  
            elif label == ' ':
                data['Engine_type'] = value  #  
            elif label == '"':
                data['Km'] = int(value.replace(',', ''))  # '
            elif label == '':
                data['Color'] = value  #  
            elif label == ' ':
                data['Prev_ownership'] = value  #  
            elif label == ' ':
                data['Curr_ownership'] = value  #  
            elif label == '':
                data['Area'] = value  # 
            elif label == '':
                data['City'] = value  # 
            elif label == ' ':
                data['Test'] = value  #  
    
    return data


#   DataFrame  
car_data_list = [parse_car_page(url) for url in ad_links]
df = pd.DataFrame(car_data_list)
#print(df)
#  DataFrame  CSV
csv_file_path = 'data/car_data.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
print(f"CSV file has been created: {csv_file_path}")
```
