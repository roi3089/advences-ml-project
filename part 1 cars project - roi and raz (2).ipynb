```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

def parse_car_page(url):
    # קבלת תגובה מהאתר ויצירת עץ DOM עם BeautifulSoup
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # מידע שנרצה לאסוף מהעמוד
    data = {
        'manufactor': 'Renault',  # יצרן הוא קבוע בדוגמה זו
        'Year': None,
        'model': None,
        'Hand': None,
        'Gear': None,
        'Engine_capacity': None,
        'Engine_type': None,
        'Prev_ownership': None,
        'Curr_ownership': None,
        'Area': None,
        'City': None,
        'Price': None,
        'Pic_num': None,
        'Cre_date': None,
        'Repub_date': None,
        'Description': None,
        'Color': None,
        'Km': None,
        'Test': None,
        'Supply_score': None  # לא ברור מה השדה הזה, לא נעשה שימוש כרגע
    }
    
    # חיפוש המודל והמחיר
    price_and_model_div = soup.find('div', class_='d-flex justify-content-between')
    if price_and_model_div:
        h2_tags = price_and_model_div.find_all('h2', class_='card-title')
        if len(h2_tags) == 2:
            data['model'] = h2_tags[0].text.strip()  # שם הדגם
            price_text = h2_tags[1].text.strip().replace('₪', '').replace(',', '').strip()
            data['Price'] = float(price_text) if price_text.isdigit() else None  # מחיר
    
    # חיפוש מספר התמונות
    pic_div = soup.find('div', itemtype='http://schema.org/ImageGallery')
    if pic_div:
        data['Pic_num'] = len(pic_div.find_all('figure'))  # מספר התמונות
    
    # חיפוש התיאור
    description_p = soup.find('p', class_='text-word-break')
    if description_p:
        data['Description'] = description_p.text.strip()  # התיאור
    
    # חיפוש תאריכי יצירה והקפצה
    dates_div = soup.find('div', class_='d-flex flex-row align-items-center justify-content-center flex-wrap')
    if dates_div:
        date_texts = dates_div.find_all('div', class_='px-3')
        for date_text in date_texts:
            text = date_text.text.strip()
            if 'תאריך יצירה' in text:
                data['Cre_date'] = datetime.strptime(text.replace('תאריך יצירה: ', '').strip(), '%d/%m/%Y')  # תאריך יצירה
            elif 'תאריך הקפצה אחרון' in text:
                data['Repub_date'] = datetime.strptime(text.replace('תאריך הקפצה אחרון: ', '').strip(), '%d/%m/%Y')  # תאריך הקפצה
    
    # חיפוש נתונים בטבלה
    rows = soup.find('table', class_='table').find_all('tr')
    for row in rows:
        cells = row.find_all('td')
        if len(cells) == 2:
            label = cells[0].text.strip()
            value = cells[1].text.strip()
            
            # השמה לשדות הנתונים המתאימים
            if label == 'שנה':
                data['Year'] = int(value)  # שנה
            elif label == 'יד':
                data['Hand'] = int(value)  # יד
            elif label == 'ת. הילוכים':
                data['Gear'] = value  # סוג הילוכים
            elif label == 'נפח':
                data['Engine_capacity'] = int(value.replace(',', ''))  # נפח המנוע
            elif label == 'סוג מנוע':
                data['Engine_type'] = value  # סוג המנוע
            elif label == 'ק"מ':
                data['Km'] = int(value.replace(',', ''))  # קילומטראז'
            elif label == 'צבע':
                data['Color'] = value  # צבע הרכב
            elif label == 'בעלות קודמת':
                data['Prev_ownership'] = value  # בעלות קודמת
            elif label == 'בעלות נוכחית':
                data['Curr_ownership'] = value  # בעלות נוכחית
            elif label == 'אזור':
                data['Area'] = value  # אזור
            elif label == 'עיר':
                data['City'] = value  # עיר
            elif label == 'טסט עד':
                data['Test'] = value  # טסט עד
    
    return data


# סידור ויצירת DataFrame מהנתונים המאוספים
car_data_list = [parse_car_page(url) for url in ad_links]
df = pd.DataFrame(car_data_list)
#print(df)
# הדפסת DataFrame לקובץ CSV
csv_file_path = 'data/car_data.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
print(f"CSV file has been created: {csv_file_path}")
```
